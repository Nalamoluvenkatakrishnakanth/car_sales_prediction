## Final processing of the data to be feed into model

```python
df = pd.get_dummies(sal_data, drop_first= True)
df.head()
```

## Feature transformation

```python
# Before transformation y values
y= df['car purchase amount']
y
```

```python
# Transforming Target Variable
y_trans = np.log(df['car purchase amount'])
y_trans
```

```python
sns.distplot(df['car purchase amount'])
```

```python
# Feature Scaling on X
x = df.drop(columns=["car purchase amount"])
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scale = scaler.fit_transform(x)
x_scale
```

```python
# Target Variable
y= df['car purchase amount']

# Independent Variables
x = df.drop(columns=['car purchase amount'])
```

# Modeling with ML algorithm and with out transformation

## Linear regression model

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scale, y_trans, test_size= 0.2, random_state=100)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

model_lr = LinearRegression()
model_lr
```

```python
#Assign linearregression model

model_lr.fit(x_train,y_train)
```

```python
y_train_pred=model_lr.predict(x_train)
y_test_pred=model_lr.predict(x_test)
```

```python
#train data
print("mae:",mean_absolute_error(y_train, y_train_pred))
print("rmse_train:",np.sqrt(mean_squared_error(y_train, y_train_pred)))
print("R2_score_train",r2_score(y_train, y_train_pred))
```

```python
#test data
print("mae:",mean_absolute_error(y_test, y_test_pred))
print("rmse_test:",np.sqrt(mean_squared_error(y_test, y_test_pred)))
print("R2_score_test",r2_score(y_test, y_test_pred))
```
